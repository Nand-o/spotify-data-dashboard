# Cleaning Data

```python
# 1. Import Library Wajib
import pandas as pd
import numpy as np

# 2. Konfigurasi Tampilan agar kolom terlihat semua
pd.set_option('display.max_columns', None)

print("Setup berhasil! Siap memproses data.")
```

```python
# Load data
df = pd.read_csv('dataset.csv')

# Cek dimensi awal (Baris, Kolom)
print(f"Ukuran Data Awal: {df.shape}")

# Cek 5 baris teratas
display(df.head())

# Cek tipe data dan data kosong
print("\nINFO DATASET:")
df.info()
```

```python
# --- 1. HAPUS KOLOM SAMPAH ---
# Cek apakah kolom 'Unnamed: 0' ada, jika ada hapus
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
    print("‚úÖ Kolom 'Unnamed: 0' berhasil dihapus.")

# --- 2. HAPUS DUPLIKAT ---
# Kita hapus duplikat berdasarkan 'track_id' agar satu lagu hanya dihitung sekali
jumlah_duplikat = df.duplicated(subset=['track_id']).sum()
print(f"‚ö†Ô∏è Ditemukan {jumlah_duplikat} lagu duplikat.")

df_clean = df.drop_duplicates(subset=['track_id'], keep='first')
print("‚úÖ Duplikat berhasil dihapus.")

# --- 3. HAPUS MISSING VALUES ---
jumlah_null = df_clean.isnull().sum().sum()
if jumlah_null > 0:
    df_clean = df_clean.dropna()
    print(f"‚úÖ Menghapus baris yang kosong (Null).")
else:
    print("‚úÖ Tidak ada data kosong.")

# --- HASIL AKHIR ---
print(f"\nUkuran Data Setelah Cleaning: {df_clean.shape}")
```

```python
# Simpan ke file CSV baru tanpa indeks
df_clean.to_csv('spotify_cleaned.csv', index=False)

print("üéâ SUKSES! File 'spotify_cleaned.csv' telah dibuat.")
print("Silakan download file tersebut dari menu Files di sebelah kiri.")
```

# Exploratory Data Analysis (EDA)
- Analisis Univariat (Distribusi Data)
- Analisis Bivariat (Matriks Korelasi)
- Analisis Hubungan Spesifik (Scatter Plot)
- Analisis Kategori (Bar Chart Genre)


```python
# Analisis Univariat (Distribusi Data)
import plotly.express as px

# Membuat Histogram Distribusi Popularitas
fig_univariat = px.histogram(df,
                   x='popularity',
                   nbins=20,
                   title='Distribusi Popularitas Lagu (Univariat)',
                   labels={'popularity':'Skor Popularitas (0-100)'},
                   color_discrete_sequence=['#1DB954'], # Warna Hijau Spotify
                   template='plotly_dark')

fig_univariat.update_layout(yaxis_title="Jumlah Lagu")
fig_univariat.show()
```

```python
# Analisis Bivariat (Matriks Korelasi)
# 1. Hitung korelasi antar kolom angka
corr_matrix = df.corr(numeric_only=True)

# 2. Visualisasi Heatmap Interaktif
fig_corr = px.imshow(corr_matrix,
                text_auto='.2f', # Menampilkan angka korelasi (2 desimal)
                aspect="auto",
                color_continuous_scale='RdBu_r', # Merah-Biru (Red-Blue reversed)
                title='Matriks Korelasi Fitur Audio (Bivariat)',
                template='plotly_dark')

fig_corr.show()
```

```python
# Analisis Hubungan Spesifik (Scatter Plot)
# Ambil sampel agar ringan
df_sample = df.sample(n=2000, random_state=42)

fig_scatter = px.scatter(df_sample,
                   x='energy',
                   y='loudness',
                   color='track_genre', # Pembeda warna berdasarkan genre
                   title='Hubungan Energy vs Loudness (Multivariat)',
                   labels={'energy': 'Energy Level', 'loudness': 'Loudness (dB)'},
                   template='plotly_dark',
                   opacity=0.7)

fig_scatter.show()
```

```python
# Analisis Kategori (Bar Chart Genre)
# Hitung rata-rata popularitas per genre, ambil top 10
genre_rank = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10).reset_index()

fig_bar = px.bar(genre_rank,
             x='popularity',
             y='track_genre',
             orientation='h', # Horizontal bar
             title='Top 10 Genre Paling Populer',
             labels={'popularity': 'Rata-rata Popularitas', 'track_genre': 'Genre'},
             color='popularity',
             color_continuous_scale='Viridis',
             template='plotly_dark')

# Membalik urutan agar yang paling populer ada di atas
fig_bar.update_layout(yaxis={'categoryorder':'total ascending'})
fig_bar.show()
```
